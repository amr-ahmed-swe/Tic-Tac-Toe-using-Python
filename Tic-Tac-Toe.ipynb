```python
board = [ [1, 2, 3,], [4, 'X', 6], [7, 8, 9] ]
number_of_moves = 1
human = 'O'
computer = 'X'

print('Hello and Welcome to Tic-Tac-Toe The Pythonic Way!!!')
print('Here\s The current status of our game board: ')
display_board(board)
print()

while number_of_moves < 9:
# THe Following code covers the user's turn each round
    print('Now it\'s time for human to make it\'s move')
    enter_move(board)
    number_of_moves += 1
    display_board(board)

    if victory_for(board, human) == True:
        print('You cushed the computer! You are a super human! Woot woot!')
        break
    else:
        print('Here\'s the current list of free squares on our board and the status: ')
        make_list_of_free_fields(board)
        print()
        display_board(board)

# THe Following code covers the computer's turn each round
    print()
    print('Now it\'s time for the computer to make it\'s move')
    draw_move(board)
    number_of_moves += 1
    display_board(board)
    print()

    if victory_for(board, computer) == True:
        print('the computer has outsmarted you once again, Human! Better luck next time!')
        break
    else:
        print('Here\'s the current list of free squares on our board and the status: ')
        make_list_of_free_fields(board)
        print()
        display_board(board)
else:
    print('We have a tie! this is a cat\'s game!!!')
print('THANK YOU FOR PLAYING!PLEASE PLAY AGAIN SOON')
 
```
